# Importing the Libraries

```python
import pandas as pd
import numpy as np
import keras
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as warn
warn.filterwarnings('ignore')
```

# Loading the data

```python
#importing from keras
from keras.datasets import mnist
```

```python
#splitting the data
(X_train,y_train), (X_test,y_test) = mnist.load_data()
```

```python
#printing the shape
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## reshaping the data

```python
#defining no of rows and cols
NUM_Rows=28
NUM_Cols=28
```

```python
#reshaping
X_train = X_train.reshape(X_train.shape[0],NUM_Rows*NUM_Cols)
X_test = X_test.reshape(X_test.shape[0],NUM_Rows*NUM_Cols)

#Scaling the data
X_train = X_train.astype('float')/255
X_test = X_test.astype('float')/255
```

```python
#now checking the max and min
np.max(X_train[0]),np.min(X_train[0])
```

```python
#let's use to_categorical
from keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

# ANN

```python
#importing libraries
from keras import *
from keras.layers import *
```

```python
#creating model_ANN1
model_ANN1 = Sequential()

#adding ip layer
model_ANN1.add(Input(shape=(NUM_Rows*NUM_Cols,)))

#adding hidden layers
model_ANN1.add(Dense(64,activation='relu'))
model_ANN1.add(Dense(32,activation='relu'))
model_ANN1.add(Dense(16,activation='relu'))

#adding op layer
model_ANN1.add(Dense(10,activation='softmax'))
```

```python
#summary
model_ANN1.summary()
```

```python
#compiling the model
model_ANN1.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#fitting the model
model_ANN1.fit(X_train,y_train,
               batch_size=10,
               epochs = 20,
               validation_data=(X_test,y_test)
)
```

```python
#checking the accuracy
_,accuracy_ANN1 = model_ANN1.evaluate(X_test,y_test)
print(accuracy_ANN1*100)
```

## now by adding Dropouts

```python
#let's add dropout after each layer

model_ANN2 = Sequential()

model_ANN2.add(Input(shape=(NUM_Rows*NUM_Cols,)))
model_ANN2.add(Dense(64,activation='relu'))
model_ANN2.add(Dropout(0.1))

model_ANN2.add(Dense(32,activation='relu'))
model_ANN2.add(Dropout(0.1))

model_ANN2.add(Dense(16,activation='relu'))
model_ANN2.add(Dropout(0.1))

model_ANN2.add(Dense(10,activation='softmax'))
```

```python
#compiling the model
model_ANN2.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#fitting the new model
model_ANN2.fit(X_train,y_train,
               batch_size=10,
               epochs=20,
               validation_data = (X_test,y_test)
               )
```

```python
#checking the accuracy
_,accuracy_ANN2 = model_ANN2.evaluate(X_test,y_test)
print(accuracy_ANN2*100)
```

#CNN

We need to import the data again since it was flattened for the ANN, but we don't need it in that format for CNN

```python
#importing again
(X_train1,y_train1), (X_test1,y_test1) = mnist.load_data()
```

```python
#printing shapes
print(X_train1.shape)
print(X_test1.shape)
print(y_train1.shape)
print(y_test1.shape)
```

```python
#Scaling it
X_train1 = X_train1.astype('float')/255
X_test1 = X_test1.astype('float')/255
```

```python
#checking the shape
X_train1.shape
```

```python
#changing the shape of X_train and X_test to avoid the dimensionality mismatch
X_train1 = np.expand_dims(X_train1,axis=-1)
X_test1 = np.expand_dims(X_test1,axis=-1)
```

```python
#before changing to catergorical
y_train1[0]
```

```python
#converting y_train and y_test to categorical
y_train1 = to_categorical(y_train1)
y_test1 = to_categorical(y_test1)
```

```python
#after changing to catergorical
y_train1[5]
```

## Building it now

```python
#Creating a CNN model
model_CNN1 = Sequential(
    [
        #CNN
        Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),

        #Bridge between CNNs and FC layers
        Flatten(),

        #Fully connected layers
        Dense(32, activation='relu'),
        Dense(10, activation='softmax')
    ]
)
```

```python
#summary of the model
model_CNN1.summary()
```

```python
#compiling our model
model_CNN1.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#fitting our model
model_CNN1.fit(X_train1,y_train1,
               batch_size=256,
               epochs=5,
               validation_data=(X_test1,y_test1)
               )
```

```python
#printing the accuracy
_,accuracy_CNN1 = model_CNN1.evaluate(X_test1,y_test1)
print(accuracy_CNN1*100)
```

## Adding more CNN layer to the model

```python
#building new model with more CNN layer
model_CNN2 = Sequential(
    [
        #CNN
        Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),

        #2nd
        Conv2D(32, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),

        #Bridge between CNNs and FC layers
        Flatten(),

        #FC layers
        Dense(32, activation='relu'),
        Dense(10, activation='softmax')
    ]
)
```

```python
#compiling new model
model_CNN2.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#fitting the model
model_CNN2.fit(X_train1,y_train1,
               batch_size=256,
               epochs=5,
               validation_data=(X_test1,y_test1)
               )
```

```python
#checking the accuracy
_,accuracy_CNN2 = model_CNN2.evaluate(X_test1,y_test1)
print(accuracy_CNN2*100)
```

```python
#so after adding 1 more CNN layer , accuracy has increased
```
